```python
import pandas as pd
from io import StringIO
import requests
import http.client
import json


conn = http.client.HTTPSConnection("uat.credresolve.com")
auth_payload = json.dumps({

    # "email": "aayush0095@gmail.com",
    # "password": "VT*Rq8&h"

    'email':'test@gmail.com',
    'password': 'tet@1234'
})
auth_headers = {
    'X-WRK-CODE': 'stage',
    'Content-Type': 'application/json',
}
conn.request("POST", "/api/v1/auth/authenticate", auth_payload, auth_headers)
res = conn.getresponse()
auth_data = json.loads(res.read())
print(auth_data)
access_token = auth_data['data']['access_token']
# print(access_token)
print("✅ Authenticated successfully.")
```
